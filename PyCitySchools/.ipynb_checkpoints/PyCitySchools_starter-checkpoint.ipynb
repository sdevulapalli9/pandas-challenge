```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Calculate the percentage of students who passed math **and** reading (% Overall Passing)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
# Total number of Schools
school_info = len(school_data_complete['school_name'].unique())
# Total number of students
students_count = len(school_data_complete['student_name'])
# Total budget
school_budget = school_data_complete['budget'].sum()
school_budget
# Average Math Score
Math_Average =  school_data_complete['math_score'].mean()
Math_Average = round(Math_Average, 2)
Math_Average
# Average Reading average
Reading_Average = school_data_complete['reading_score'].mean()
Reading_Average = round(Reading_Average, 2)
Reading_Average
# Math Score greater than 70
Math_above_70 =school_data_complete[(school_data_complete["math_score"] >= 70)].count()['student_name']
Math_above_70
# Math_score_Avg
Math_score_Avg_percentage = Math_above_70/students_count*100
Math_score_Avg_percentage = round(Math_score_Avg_percentage,2)
Math_score_Avg_percentage

# Reading Score greater than 70
reading_above_70 = school_data_complete[(school_data_complete["reading_score"] >= 70)].count()['student_name']
#Reading_Score_Avg
Reading_score_Avg_percentage = reading_above_70/students_count*100
Reading_score_Avg_percentage = round(Reading_score_Avg_percentage,2)
Reading_score_Avg_percentage

Overall_Average = school_data_complete[(school_data_complete["math_score"] >= 70) 
                                                  & (school_data_complete["reading_score"] >= 70)].count()["student_name"]
Overall_Average_Percentage = (Overall_Average/students_count)*100
Overall_Average_Percentage = round(Overall_Average_Percentage,2)
district_summary = pd.DataFrame({"Total Schools":[school_info],
                                "Total students":[students_count],
                                "Total Budget":[school_budget],
                                "Average Math Score":[Math_Average],
                                 "Average Reading Score":[Reading_Average],
                                 "Math %":[Math_score_Avg_percentage],
                                  "Reading %":[Reading_score_Avg_percentage],
                                   "Overall %":[Overall_Average_Percentage]})
district_summary
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * % Overall Passing (The percentage of students that passed math **and** reading.)
  
* Create a dataframe to hold the above results

```python
school_info = school_data.set_index(["school_name"])["type"]
Total_students = school_data_complete["school_name"].value_counts()
Total_school_budget = school_data_complete.groupby(["school_name"])["budget"].mean()
Total_school_budget
per_student_budget = Total_school_budget/Total_students
per_student_budget
Avg_math_score = school_data_complete.groupby(["school_name"])["math_score"].mean()
Avg_math_score
Avg_reading_score = school_data_complete.groupby(["school_name"])["reading_score"].mean()
Avg_reading_score
Math_passing = school_data_complete[(school_data_complete["math_score"] >= 70)].groupby(["school_name"])["student_name"].count()
Math_passing_Percentage = Math_passing/Total_students*100
Math_passing_Percentage
Read_passing = school_data_complete[(school_data_complete["reading_score"] >= 70)].groupby(["school_name"])["student_name"].count()
Read_passing_Percentage = Read_passing/Total_students*100
Read_passing_Percentage
Overall_passing = school_data_complete[(school_data_complete["math_score"] >= 70) 
                                                  & (school_data_complete["reading_score"] >= 70)].groupby(["school_name"])["student_name"].count()
Overall_passing_Percentage  = Overall_passing/Total_students*100
Overall_passing_Percentage
school_info
```

```python
school_summary = pd.DataFrame({"school Type":school_info,
                               "Total students":Total_students,
                               "Total_school_budget":Total_school_budget,
                              "Per_Student_Budget":per_student_budget,
                              "Avg_math_score":Avg_math_score,
                               "Avg_reading_score":Avg_reading_score,
                              "Math_Passing_Percentage":Math_passing_Percentage,
                              "Read_Passing_Percentage":Read_passing_Percentage,
                              "Overall_Passing_Percentage":Overall_passing_Percentage})
school_summary
```

## Top Performing Schools (By % Overall Passing)

```python
top_performing = school_summmary.sort_values(["Overall_Passing_Percentage"], ascending=False)
top_performing
```

## Math Scores by Grade

* Create a table that lists the average Reading Score for students of each grade level (9th, 10th, 11th, 12th) at each school.

  * Create a pandas series for each grade. Hint: use a conditional statement.
  
  * Group each series by school
  
  * Combine the series into a dataframe
  
  * Optional: give the displayed data cleaner formatting

```python
ninth_grade = school_data_complete[(school_data_complete["grade"]=="9th")].groupby("school_name")["math_score"].mean()
Tenth_grade = school_data_complete[(school_data_complete["grade"]=="10th")].groupby("school_name")["math_score"].mean()
Eleventh_grade = school_data_complete[(school_data_complete["grade"]=="11th")].groupby("school_name")["math_score"].mean()
Twelth_grade = school_data_complete[(school_data_complete["grade"]=="12th")].groupby("school_name")["math_score"].mean()
math_score_grade = pd.DataFrame({"9th":ninth_grade,
                                "10th":Tenth_grade,
                                "11th":Eleventh_grade,
                                "12th":Twelth_grade})
math_score_grade
```

## Reading Score by Grade

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
bin_1 = [0,585,630,645,675]
spending_group = ["<$584", "$585-629", "$630-644", "$645-675"]
school_summary
```

```python
school_summary["spending Ranges(per student)"] = pd.cut(per_student_budget,bin_1,labels=spending_group)
school_summary
spending_math_scores = school_summary.groupby(["spending Ranges(per student)"])["Avg_math_score"].mean()
spending_reading_scores = school_summary.groupby(["spending Ranges(per student)"])["Avg_reading_score"].mean()
spending_math_passing = school_summary.groupby(["spending Ranges(per student)"])["Math_Passing_Percentage"].mean()
spending_read_passing = school_summary.groupby(["spending Ranges(per student)"])["Read_Passing_Percentage"].mean()
spending_overall_passing = school_summary.groupby(["spending Ranges(per student)"])["Overall_Passing_Percentage"].mean()
spending_summary = pd.DataFrame({"Ave Math Score ":spending_math_scores,
                                "Avg Reading Score":spending_reading_scores,
                                "% Passing Math":spending_math_passing,
                                "% Passing Reading":spending_read_passing,
                                "% Overall Passing":spending_overall_passing})
spending_summary
```

## Scores by School Size

* Perform the same operations as above, based on school size.

```python
bin_2 = [0, 1000, 2000, 5000]
group_names = ["Small (<1000)", "Medium (1000-2000)", "Large (2000-5000)"]
school_summary["School Size"] = pd.cut(school_summary["Total students"], bin_2, labels=group_names)
school_summary
size_math_score = school_summary.groupby(["School Size"])["Avg_math_score"].mean()
size_read_score = school_summary.groupby(["School Size"])["Avg_reading_score"].mean()
size_math_passing = school_summary.groupby(["School Size"])["Math_Passing_Percentage"].mean()
size_read_passing= school_summary.groupby(["School Size"])["Read_Passing_Percentage"].mean()
size_overall_passing = school_summary.groupby(["School Size"])["Overall_Passing_Percentage"].mean()
school_size_summary = pd.DataFrame({"Average Math Score":size_math_score,
                                   "Average Reading Score": size_read_score,
                             "% Passing Math": size_math_passing,
                             "% Passing Reading": size_overall_passing,
                             "% Overall Passing": size_overall_passing}) 
school_size_summary
```
